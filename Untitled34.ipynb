```python
# agora vou iniciar a limpeza dos dados
df_cleaned = df.drop_duplicates()
```

```python
valores_ausentes = df.isnull().sum()
print("Valores ausentes por coluna:\n", valores_ausentes)
```

```python
df_cleaned = df.dropna()
```

```python
valores_ausentes_pos_remocao = df_cleaned.isnull().sum()
print("Valores ausentes após a remoção:\n", valores_ausentes_pos_remocao)
```

```python
# Identificar as colunas numéricas
colunas_numericas = df_cleaned.select_dtypes(include=['int64', 'float64']).columns
print("Colunas numéricas:\n", colunas_numericas)
```

```python


# Criar o objeto MinMaxScaler
scaler = MinMaxScaler()

# Aplicar a normalização nas colunas numéricas
df_cleaned[colunas_numericas] = scaler.fit_transform(df_cleaned[colunas_numericas])

# Verificar o resultado da normalização
df_cleaned[colunas_numericas].head()
```

```python
#Conversão de Colunas Categóricas Vou aplicar o One-Hot Encoding nas variáveis categóricas, transformando cada categoria em uma nova coluna binária, segue o passo a passo pra fazer.
```

```python
# Identificar colunas categóricas
colunas_categoricas = df_cleaned.select_dtypes(include=['object']).columns
print("Colunas categóricas:\n", colunas_categoricas)
```

```python
# Aplicar One-Hot Encoding nas colunas categóricas
df_encoded = pd.get_dummies(df_cleaned, columns=colunas_categoricas, drop_first=True)
```

```python
# Verificar o resultado após a conversão
df_encoded.head()
```

```python
#Separação das Variáveis X e Y, Agora vou separar as variáveis X (variáveis independentes) e Y (variável dependente, que neste caso será Attrition_Flag).
```

```python
# Verificar os nomes das colunas após o One-Hot Encoding
df_encoded.columns
```

```python
# Definir Y (variável dependente) como o inverso de 'Attrition_Flag_Existing Customer'
# 1 será churn (cliente que saiu), e 0 será não churn (cliente que ficou)
Y = 1 - df_encoded['Attrition_Flag_Existing Customer']

# Remover colunas desnecessárias de X, como 'CLIENTNUM' e a coluna de churn 'Attrition_Flag_Existing Customer'
X = df_encoded.drop(['CLIENTNUM', 'Attrition_Flag_Existing Customer'], axis=1)

# Verificar as dimensões de X e Y
print("Dimensões de X:", X.shape)
print("Dimensões de Y:", Y.shape)
```

```python

# Dividir o conjunto de dados em treinamento e teste (80% treino, 20% teste)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Verificar as dimensões após a divisão
print("Dimensões do conjunto de treinamento:", X_train.shape)
print("Dimensões do conjunto de teste:", X_test.shape)
```

```python
# Treinamento e Avaliação de Modelos de Machine Learning
#Agora, vou testar um algoritmo de classificação. Vou começar com Random Forest, um modelo de árvore de decisão que geralmente apresenta bons resultados.
```

```python

# Inicializar o modelo Random Forest
rf_model = RandomForestClassifier(random_state=42)

# Treinar o modelo
rf_model.fit(X_train, Y_train)

# Fazer previsões no conjunto de teste
Y_pred = rf_model.predict(X_test)

# Avaliar o desempenho do modelo
print("Acurácia:", accuracy_score(Y_test, Y_pred))
print("\nMatriz de Confusão:\n", confusion_matrix(Y_test, Y_pred))
print("\nRelatório de Classificação:\n", classification_report(Y_test, Y_pred))
```

```python
# Ajustando hiperparâmetros para evitar overfitting
rf_model_adjusted = RandomForestClassifier(max_depth=5, min_samples_leaf=10, random_state=42)

# Treinar o modelo ajustado
rf_model_adjusted.fit(X_train, Y_train)

# Fazer previsões no conjunto de teste
Y_pred_adjusted = rf_model_adjusted.predict(X_test)

# Avaliar o desempenho do modelo ajustado
print("Acurácia Ajustada:", accuracy_score(Y_test, Y_pred_adjusted))
print("\nMatriz de Confusão Ajustada:\n", confusion_matrix(Y_test, Y_pred_adjusted))
print("\nRelatório de Classificação Ajustada:\n", classification_report(Y_test, Y_pred_adjusted))
```

```python
#Aplicar validação cruzada (Cross-Validation), Para avaliar o modelo de maneira mais confiável, podemos usar a técnica de validação cruzada (K-Fold Cross Validation):
```

```python
from sklearn.model_selection import cross_val_score

# Aplicar validação cruzada com 5 folds
cross_val_scores = cross_val_score(rf_model_adjusted, X, Y, cv=5)

# Exibir a acurácia média
print("Acurácia média da validação cruzada:", cross_val_scores.mean())
```

```python
from sklearn.linear_model import LogisticRegression

# Inicializar o modelo de Regressão Logística
logreg = LogisticRegression(random_state=42)

# Treinar o modelo
logreg.fit(X_train, Y_train)

# Fazer previsões no conjunto de teste
Y_pred_logreg = logreg.predict(X_test)

# Avaliar o desempenho
print("Acurácia Regressão Logística:", accuracy_score(Y_test, Y_pred_logreg))
print("\nMatriz de Confusão:\n", confusion_matrix(Y_test, Y_pred_logreg))
print("\nRelatório de Classificação:\n", classification_report(Y_test, Y_pred_logreg))
```

```python
from sklearn.model_selection import GridSearchCV

# Definir o grid de hiperparâmetros
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Inicializar o GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Treinar com a busca de hiperparâmetros
grid_search.fit(X_train, Y_train)

# Exibir os melhores hiperparâmetros
print("Melhores parâmetros:", grid_search.best_params_)
```
